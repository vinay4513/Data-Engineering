#Target Table format

```
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```
target_table = sample.select(
    row_number().over(Window.orderBy(monotonically_increasing_id())).alias("Abacus_Record_ID"),
    expr("substr(EMPLOYEE_ID,-3,3) || row_number() over(partition by EMPLOYEE_ID order by 1 ) || GENDER || substr(Date(DOB),-2,2)").alias("Abacus_Member_ID"),
    sample["MEMBER_ID"].alias("Member_ID"),
    sample["EMPLOYEE_ID"].alias("Subscriber_ID"),
    sample["FIRST_NAME"].alias("Member_First_Name"),
    sample["LAST_NAME"].alias("Member_Last_Name"),
    sample["MIDDLE_NAME"].alias("Member_Middle_Name"),

    lit(None).alias("Member_Prefix_Name"),
    lit(None).alias("Member_Suffix_Name"),

    sample["Gender_Description"].alias("Member_Gender"),
    date_format(col("DOB"), "MM/dd/yyyy").alias("Member_Date_of_Birth"),
    sample["Relationship_Description"].alias("Member_Relationship_Description"),
    sample["Relationship_Code"].alias("Member_Relationship_Code"),
    sample["PERSON_CODE"].alias("Member_Person_Code").cast("integer"),
    sample["ADDRESS_1"].alias("Member_Address_Line_1"),
    sample["ADDRESS_2"].alias("Member_Address_Line_2"),
    sample["CITY"].alias("Member_City"),
    sample["STATE"].alias("Member_State"),
    sample["COUNTY"].alias("Member_County"),
    lit("USA").alias("Member_Country"),

    lit(None).alias("Member_Home_Phone"),
    lit(None).alias("Member_Work_Phone"),
    lit(None).alias("Member_Mobile_Phone"),
    lit(None).alias("Member_Email"),
    lit(None).alias("Member_Is_Deceased"),
    lit(None).alias("Member_Date_of_Death"),
    lit(None).alias("Member_Deceased_Reason"),

    sample["ZIP"].alias("Member_Postal_Code"),
    sample["GROUP_ID"].alias("Enrollment_Group_ID"),
    sample["GROUP_NAME"].alias("Enrollment_Group_Name"),

    lit(None).alias("Enrollment_SubGroup_ID"),
    lit(None).alias("Enrollment_SubGroup_Name"),

    sample["Coverage_ID"].alias("Enrollment_Coverage_Code"),
    sample["Coverage_Description"].alias("Enrollment_Coverage_Description"),
    sample["PLAN_ID"].alias("Enrollment_Plan_ID"),
    sample["Plan Name"].alias("Enrollment_Plan_Name"),
    sample["BENEFIT_TYPE"].alias("Enrollment_Plan_Coverage"),
    when(col("BENEFIT_TYPE").contains("Medical"),date_format(col("EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Medical_Effective_Date"),
    when(col("BENEFIT_TYPE").contains("Medical"),date_format(col("TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Medical_Termination_Date"),
    when(col("BENEFIT_TYPE").contains("Dental"),date_format(col("EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Dental_Effective_Date"),
    when(col("BENEFIT_TYPE").contains("Dental"),date_format(col("TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Dental_Termination_Date"),
    when(col("BENEFIT_TYPE").contains("Vision"),date_format(col("EFFECTIVE_DATE"),"MM/dd/yyyy")).alias("Enrollment_Vision_Effective_Date"),
    when(col("BENEFIT_TYPE").contains("Vision"),date_format(col("TERMINATION_DATE"),"MM/dd/yyyy")).alias("Enrollment_Vision_Termination_Date"),
    sample["VENDOR"].alias("Enrollment_Vendor_Name"),
    lit("Member Enrollment").alias("Souce_File_Name"),
    date_format(current_date(),"MM/dd/yyyy").alias("File_Ingestion_Date")
    



)
```
